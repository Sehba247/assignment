- ANS 3 - In hypothesis testing, the significance level (α) is the probability of rejecting the null hypothesis when it is actually true. It essentially defines the threshold of evidence required to reject the null hypothesis. Why is it important? It controls the risk of making a Type I error (false positive), ensuring the validity and reliability of the findings.

By setting a significance level (e.g., 0.05, 0.01), researchers define how much risk they're willing to take of making a Type I error.
Threshold for Rejection:
The significance level acts as the threshold. If the p-value (the probability of observing the data if the null hypothesis were true) is less than the significance level, the null hypothesis is rejected.
A low significance level (e.g., 0.01) indicates that a researcher requires stronger evidence before rejecting the null hypothesis. This ensures that conclusions are more likely to be correct and avoids false positives.

-ANS -12 -A t-test is a statistical hypothesis test used to determine if there's a significant difference between the means of two groups.
How it's used in hypothesis testing:
1. State the null and alternative hypotheses:
The null hypothesis assumes no significant difference between the means, while the alternative hypothesis suggests there is a significant difference.
2. Choose a t-test type:
There are different types of t-tests, including one-sample, independent samples, and paired samples t-tests. The choice depends on the study design and the relationship between the groups being compared.
3. Calculate the t-statistic:
The t-statistic is a measure of how much the sample means differ, considering the variability within the groups.
4. Determine the degrees of freedom:
Degrees of freedom represent the number of independent pieces of information used to calculate the t-statistic.
5. Compare the t-statistic to a critical value or p-value:
The critical value is determined based on the chosen significance level (alpha) and the degrees of freedom. A p-value represents the probability of observing the data if the null hypothesis is true.
6. Make a decision:
If the calculated t-statistic is greater than the critical value or the p-value is less than the significance level, the null hypothesis is rejected, indicating a statistically significant difference between the means.

- ANS -14 - A confidence interval is a range of values within which there is a specified probability that the true population parameter lies, based on sample data. It's used to estimate an unknown parameter, like a population mean, with a certain degree of confidence, often 95% or 99%.
Here's how it's used to interpret statistical results:
Estimating Population Parameters:
Instead of just providing a point estimate (like a single average), confidence intervals offer a range of plausible values for the true population parameter.
Quantifying Uncertainty:
They indicate the degree of uncertainty or precision in the estimate, with narrower intervals implying more precise estimates.
Hypothesis Testing:
Confidence intervals can be used to test hypotheses. For example, if a confidence interval for the difference between two groups does not include zero, it suggests a statistically significant difference between the groups.
Interpreting Results:

Understanding Precision:

- ANS -15 -The margin of error in a confidence interval quantifies the potential difference between a sample statistic (e.g., a sample mean) and the true population parameter. It essentially tells you how much your sample result might vary from the real population value. The margin of error is half the length of the confidence interval.
How the Margin of Error Affects the Confidence Interval:
Wider Interval:
A larger margin of error results in a wider confidence interval. This indicates that there's a greater range within which the true population parameter is likely to fall, making the estimate less precise.
Narrows Interval:
A smaller margin of error leads to a narrower confidence interval. This means you are more confident that the true population parameter is within a tighter range, increasing the precision of your estimate.
Confidence Level:
The confidence level (e.g., 95%, 99%) also influences the margin of error. A higher confidence level generally requires a wider interval, meaning a larger margin of error.
Sample Size:
A larger sample size typically leads to a smaller margin of error and, therefore, a narrower confidence interval. This is because a larger sample provides more information about the population, reducing uncertainty.
Population Variability:
If there is more variability in the population (i.e., the data points are spread out more), the margin of error will tend to be larger, resulting in a wider confidence interval.
In essence, the margin of error and the confidence interval are directly related: a larger margin of error results in a larger confidence interval, and vice-versa. Understanding this relationship is crucial for interpreting confidence intervals and assessing the precision of your estimates.

```python
#ANS3
import numpy as np
from scipy.stats import bernoulli

def bernoulli_pdf(x, p):
    """
    Calculates the probability mass function (PDF) of a Bernoulli distribution.

    Args:
        x: The outcome (0 or 1) to evaluate the PDF at.
        p: The probability of success (probability of x=1).

    Returns:
        The probability of the outcome x given the probability of success p.
    """
    return bernoulli.pmf(x, p)

# Example usage:
p = 0.6  # Probability of success
x = 1     # Outcome (0 or 1)
probability = bernoulli_pdf(x, p)
print(f"Probability of {x} with p={p}: {probability}")

x = 0
probability = bernoulli_pdf(x, p)
print(f"Probability of {x} with p={p}: {probability}")
```

```python
#ANS6
import numpy as np
import matplotlib.pyplot as plt

def discrete_uniform_cdf(low, high, x):
    """
    Calculates the CDF of a discrete uniform distribution.

    Args:
        low (int): The lowest value in the distribution.
        high (int): The highest value in the distribution.
        x (array-like): The values at which to evaluate the CDF.

    Returns:
        array-like: The CDF values corresponding to x.
    """
    if not isinstance(low, int) or not isinstance(high, int):
        raise ValueError("low and high must be integers")
    if low >= high:
        raise ValueError("low must be less than high")

    x = np.array(x)
    cdf = np.zeros_like(x, dtype=float)

    for i in range(len(x)):
        if x[i] < low:
            cdf[i] = 0
        elif x[i] >= high:
            cdf[i] = 1
        else:
            cdf[i] = (np.floor(x[i]) - low + 1) / (high - low)
    return cdf

def plot_cdf(low, high):
    """
    Plots the CDF of a discrete uniform distribution.

    Args:
        low (int): The lowest value in the distribution.
        high (int): The highest value in the distribution.
    """
    x = np.arange(low - 2, high + 2, 0.1)
    cdf = discrete_uniform_cdf(low, high, x)

    plt.step(x, cdf, where='post')
    plt.title(f'CDF of Discrete Uniform Distribution [{low}, {high}]')
    plt.xlabel('x')
    plt.ylabel('CDF(x)')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    low = 1
    high = 6
    plot_cdf(low, high)
```

```python
#ANS10
import numpy as np
import matplotlib.pyplot as plt

def clt_simulation(data, sample_size, num_samples):
    """
    Demonstrates the Central Limit Theorem.

    Args:
        data: A list or numpy array representing the population data.
        sample_size: The size of each sample.
        num_samples: The number of samples to draw.

    Returns:
        A list of sample means.
    """
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(data, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))
    return sample_means

if __name__ == '__main__':
    # Example using a non-normal distribution (e.g., exponential)
    population_data = np.random.exponential(scale=2.0, size=1000)
    sample_size = 30
    num_samples = 500

    sample_means = clt_simulation(population_data, sample_size, num_samples)

    # Plotting the distribution of sample means
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='g')
    plt.title(f'Distribution of Sample Means (Sample Size: {sample_size}, Number of Samples: {num_samples})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')
    plt.show()
```

```python
#ANS14
import numpy as np
from scipy.stats import norm

def calculate_z_score(data_point, data):
    """
    Calculate the Z-score for a given data point within a dataset.

    Args:
        data_point (float): The individual data point.
        data (list or numpy.ndarray): The dataset.

    Returns:
        float: The calculated Z-score.
    """
    mean = np.mean(data)
    std_dev = np.std(data)
    z_score = (data_point - mean) / std_dev
    return z_score

def compare_to_standard_normal(z_score):
    """
    Compare the Z-score to the standard normal distribution.

    Args:
        z_score (float): The Z-score to compare.

    Returns:
         float: The probability of observing a value as extreme as or more extreme than the z-score.
    """
    p_value = 2 * (1 - norm.cdf(abs(z_score)))
    return p_value

if __name__ == "__main__":
    data = [70, 85, 92, 65, 78, 88, 95, 75, 80, 90]
    data_point = 82

    z_score = calculate_z_score(data_point, data)
    print(f"Z-score for {data_point}: {z_score:.2f}")

    p_value = compare_to_standard_normal(z_score)
    print(f"P-value: {p_value:.4f}")

    alpha = 0.05
    if p_value < alpha:
        print(f"The result is statistically significant (p < {alpha})")
    else:
        print(f"The result is not statistically significant (p >= {alpha})")
```

```python
#ANS15
import numpy as np
from scipy import stats

# Sample data
data = np.array([88, 92, 94, 96, 98, 102, 104, 106, 108, 110])

# Null hypothesis: population mean = 100
# Alternative hypothesis: population mean != 100 (two-tailed test)
population_mean = 100
alpha = 0.05  # Significance level

# Calculate the z-statistic
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
sample_size = len(data)
z_statistic = (sample_mean - population_mean) / (sample_std / np.sqrt(sample_size))

# Calculate the p-value
p_value = 2 * stats.norm.cdf(-np.abs(z_statistic))

# Compare p-value to alpha
if p_value < alpha:
    print("Reject the null hypothesis")
    print(f"Z-statistic: {z_statistic:.2f}, p-value: {p_value:.3f}")
else:
    print("Fail to reject the null hypothesis")
    print(f"Z-statistic: {z_statistic:.2f}, p-value: {p_value:.3f}")
```

```python
#PRACTICAL PART-2
#ANS1
import numpy as np
from scipy import stats

def perform_z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha=0.05):
  """
  Performs a Z-test for comparing a sample mean to a known population mean.

  Args:
    sample_mean (float): The mean of the sample.
    population_mean (float): The known mean of the population.
    population_std_dev (float): The known standard deviation of the population.
    sample_size (int): The size of the sample.
    alpha (float, optional): The significance level. Defaults to 0.05.

  Returns:
    tuple: A tuple containing:
        - float: The calculated Z-statistic.
        - float: The calculated p-value.
        - str: The interpretation of the results.
  """
  # Calculate the standard error of the mean
  std_error = population_std_dev / np.sqrt(sample_size)

  # Calculate the Z-statistic
  z_statistic = (sample_mean - population_mean) / std_error

  # Calculate the p-value (two-tailed test)
  # We use 2 * (1 - cdf(abs(z))) for a two-tailed test
  p_value = 2 * (1 - stats.norm.cdf(np.abs(z_statistic)))

  # Interpret the results
  if p_value < alpha:
    interpretation = f"Reject the null hypothesis (p < {alpha}). There is sufficient evidence to suggest that the sample mean is significantly different from the population mean."
  else:
    interpretation = f"Fail to reject the null hypothesis (p >= {alpha}). There is not enough evidence to suggest that the sample mean is significantly different from the population mean."

  return z_statistic, p_value, interpretation

# Example Usage:
# Let's assume a population of students has a known average test score of 75 with a standard deviation of 10.
# We take a sample of 30 students from a different program and find their average score is 78.
# We want to know if the average score of this sample is significantly different from the population mean.

population_mean = 75
population_std_dev = 10
sample_mean = 78
sample_size = 30
alpha = 0.05

z_stat, p_val, result_interpretation = perform_z_test(sample_mean, population_mean, population_std_dev, sample_size, alpha)

print(f"Sample Mean: {sample_mean}")
print(f"Population Mean: {population_mean}")
print(f"Population Standard Deviation: {population_std_dev}")
print(f"Sample Size: {sample_size}")
print(f"Significance Level (alpha): {alpha}")
print("-" * 30)
print(f"Calculated Z-statistic: {z_stat:.4f}")
print(f"Calculated p-value: {p_val:.4f}")
print("\nInterpretation:")
print(result_interpretation)

# Another Example:
# Assume the average height of adult males in a city is 175 cm with a standard deviation of 7 cm.
# A sample of 50 adult males from a specific neighborhood has an average height of 173 cm.
# Is the average height in this neighborhood significantly different?

population_mean_height = 175
population_std_dev_height = 7
sample_mean_height = 173
sample_size_height = 50
alpha_height = 0.01 # Using a stricter significance level

z_stat_height, p_val_height, result_interpretation_height = perform_z_test(
    sample_mean_height, population_mean_height, population_std_dev_height, sample_size_height, alpha_height
)

print("\n" + "=" * 30 + "\n")
print(f"Sample Mean Height: {sample_mean_height}")
print(f"Population Mean Height: {population_mean_height}")
print(f"Population Standard Deviation Height: {population_std_dev_height}")
print(f"Sample Size Height: {sample_size_height}")
print(f"Significance Level (alpha): {alpha_height}")
print("-" * 30)
print(f"Calculated Z-statistic: {z_stat_height:.4f}")
print(f"Calculated p-value: {p_val_height:.4f}")
print("\nInterpretation:")
print(result_interpretation_height)
```

```python
#ANS3
import numpy as np
from scipy.stats import norm

def one_sample_ztest(sample, pop_mean, pop_std, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample Z-test.

    Args:
        sample (list or numpy.ndarray): The sample data.
        pop_mean (float): The population mean.
        pop_std (float): The population standard deviation.
        alpha (float, optional): Significance level. Defaults to 0.05.
        alternative (str, optional): 'two-sided', 'less', or 'greater'.
            Specifies the alternative hypothesis. Defaults to 'two-sided'.

    Returns:
        tuple: A tuple containing the z-statistic and the p-value.
    """
    n = len(sample)
    sample_mean = np.mean(sample)
    z_stat = (sample_mean - pop_mean) / (pop_std / np.sqrt(n))

    if alternative == 'two-sided':
        p_value = 2 * norm.cdf(-np.abs(z_stat))
    elif alternative == 'less':
        p_value = norm.cdf(z_stat)
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_stat)
    else:
        raise ValueError("alternative must be 'two-sided', 'less', or 'greater'")

    return z_stat, p_value

if __name__ == '__main__':
    # Example usage
    sample_data = [78, 85, 92, 88, 95, 80, 84, 89, 90, 83]
    population_mean = 85
    population_std = 5

    z_statistic, p_value = one_sample_ztest(sample_data, population_mean, population_std)
    print("Z-statistic:", z_statistic)
    print("P-value:", p_value)

    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject the null hypothesis")
```

```python
#ans5
# Type I and Type II Errors Visualization

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_hypothesis_errors(mu_null, sigma, alpha, mu_alt=None, sample_size=30):
    """
    Calculates and visualizes Type I and Type II errors during hypothesis testing.

    Args:
        mu_null (float): The mean under the null hypothesis.
        sigma (float): The population standard deviation.
        alpha (float): The significance level (Type I error rate).
        mu_alt (float, optional): The mean under the alternative hypothesis.
                                   Required to visualize Type II error.
        sample_size (int, optional): The size of the sample. Defaults to 30.
    """

    # Calculate the standard error of the mean
    std_error = sigma / np.sqrt(sample_size)

    # Define the standard normal distribution for the null hypothesis
    x = np.linspace(mu_null - 4 * std_error, mu_null + 4 * std_error, 200)
    pdf_null = norm.pdf(x, mu_null, std_error)

    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf_null, label=f'Null Hypothesis (μ = {mu_null})')

    # Calculate the critical value for a two-tailed test
    # This is the value on the x-axis that separates the rejection region
    critical_value_upper = norm.ppf(1 - alpha / 2, loc=mu_null, scale=std_error)
    critical_value_lower = norm.ppf(alpha / 2, loc=mu_null, scale=std_error)

    # Shade the Type I error region (alpha)
    x_type1_upper = np.linspace(critical_value_upper, mu_null + 4 * std_error, 100)
    y_type1_upper = norm.pdf(x_type1_upper, mu_null, std_error)
    plt.fill_between(x_type1_upper, y_type1_upper, color='red', alpha=0.3, label=f'Type I Error (α = {alpha})')

    x_type1_lower = np.linspace(mu_null - 4 * std_error, critical_value_lower, 100)
    y_type1_lower = norm.pdf(x_type1_lower, mu_null, std_error)
    plt.fill_between(x_type1_lower, y_type1_lower, color='red', alpha=0.3)

    # Add critical value lines
    plt.axvline(critical_value_upper, color='red', linestyle='--', label=f'Critical Value ({critical_value_upper:.2f})')
    plt.axvline(critical_value_lower, color='red', linestyle='--')

    # Visualize Type II error if alternative mean is provided
    if mu_alt is not None:
        # Define the distribution for the alternative hypothesis
        pdf_alt = norm.pdf(x, mu_alt, std_error)
        plt.plot(x, pdf_alt, label=f'Alternative Hypothesis (μ = {mu_alt})', linestyle='--')

        # Calculate the Type II error region (beta)
        # This is the area under the alternative distribution within the acceptance region of the null
        # The acceptance region for the null is between the lower and upper critical values
        beta = norm.cdf(critical_value_upper, loc=mu_alt, scale=std_error) - norm.cdf(critical_value_lower, loc=mu_alt, scale=std_error)

        x_type2 = np.linspace(critical_value_lower, critical_value_upper, 100)
        y_type2 = norm.pdf(x_type2, mu_alt, std_error)
        plt.fill_between(x_type2, y_type2, color='blue', alpha=0.3, label=f'Type II Error (β = {beta:.4f})')

        plt.title('Type I and Type II Errors Visualization')
    else:
         plt.title('Type I Error Visualization')


    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example Usage:
# Visualize Type I error only
plot_hypothesis_errors(mu_null=50, sigma=10, alpha=0.05, sample_size=100)

# Visualize Type I and Type II errors
plot_hypothesis_errors(mu_null=50, sigma=10, alpha=0.05, mu_alt=53, sample_size=100)
```

```python
#ans8
import numpy as np
from scipy import stats

def compare_z_t_tests(pop_mean=50, pop_std=10, sample_size_large=100, sample_size_small=15, alpha=0.05):
    """
    Simulates data and performs Z-test and T-test, then compares the results.

    Args:
        pop_mean (float, optional): Population mean. Defaults to 50.
        pop_std (float, optional): Population standard deviation. Defaults to 10.
        sample_size_large (int, optional): Size for the large sample (Z-test). Defaults to 100.
        sample_size_small (int, optional): Size for the small sample (T-test). Defaults to 15.
        alpha (float, optional): Significance level. Defaults to 0.05.
    """

    print("--- Comparing Z-test and T-test ---")
    print(f"Population Mean: {pop_mean}, Population Std Dev: {pop_std}, Alpha: {alpha}")

    # --- Simulate Data ---
    # For Z-test (requires known population std dev and typically large sample)
    np.random.seed(42) # for reproducibility
    large_sample = np.random.normal(loc=pop_mean + 2, scale=pop_std, size=sample_size_large) # Simulate a sample slightly different from pop_mean

    # For T-test (used when population std dev is unknown or sample size is small)
    small_sample = np.random.normal(loc=pop_mean + 2, scale=pop_std, size=sample_size_small) # Simulate a sample slightly different from pop_mean

    print(f"\nSimulated Large Sample Size for Z-test: {sample_size_large}")
    print(f"Simulated Small Sample Size for T-test: {sample_size_small}")

    # --- Perform Z-test (assuming population std dev is known) ---
    print("\n--- Performing Z-test ---")
    sample_mean_large = np.mean(large_sample)
    # Use population standard deviation for Z-test
    z_statistic = (sample_mean_large - pop_mean) / (pop_std / np.sqrt(sample_size_large))
    p_value_z = 2 * (1 - stats.norm.cdf(np.abs(z_statistic))) # Two-tailed p-value

    print(f"Large Sample Mean: {sample_mean_large:.2f}")
    print(f"Z-Statistic: {z_statistic:.4f}")
    print(f"P-value (Z-test): {p_value_z:.4f}")

    if p_value_z < alpha:
        print("Z-test Result: Reject the null hypothesis.")
    else:
        print("Z-test Result: Fail to reject the null hypothesis.")

    # --- Perform T-test (assuming population std dev is unknown) ---
    print("\n--- Performing T-test ---")
    sample_mean_small = np.mean(small_sample)
    # Use sample standard deviation for T-test (scipy.stats.ttest_1samp handles this)
    t_statistic, p_value_t = stats.ttest_1samp(small_sample, pop_mean)

    print(f"Small Sample Mean: {sample_mean_small:.2f}")
    print(f"T-Statistic: {t_statistic:.4f}")
    print(f"P-value (T-test): {p_value_t:.4f}")

    if p_value_t < alpha:
        print("T-test Result: Reject the null hypothesis.")
    else:
        print("T-test Result: Fail to reject the null hypothesis.")

    # --- Comparison ---
    print("\n--- Comparison ---")
    print(f"Z-test P-value: {p_value_z:.4f}")
    print(f"T-test P-value: {p_value_t:.4f}")

    if sample_size_small < 30 and pop_std is None:
         print("\nNote: T-test is more appropriate for the small sample size and unknown population standard deviation.")
    elif sample_size_large >= 30 and pop_std is not None:
         print("\nNote: Z-test is appropriate for the large sample size and known population standard deviation.")
    elif sample_size_large >= 30 and pop_std is None:
         print("\nNote: T-test would generally be preferred even with a large sample if population standard deviation is unknown.")
    elif sample_size_small < 30 and pop_std is not None:
        print("\nNote: While Z-test could technically be used with a small sample and known population std dev, T-test is more robust, especially if the population isn't perfectly normal.")


# Run the comparison
compare_z_t_tests()

# Example with a smaller difference in sample means
# compare_z_t_tests(pop_mean=50, pop_std=10, sample_size_large=100, sample_size_small=15, alpha=0.05)
```

```python
#ans9
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence_level=0.95):
    """
    Calculates the confidence interval for a sample mean.

    Args:
        data: A list or numpy array of sample data.
        confidence_level: The desired confidence level (e.g., 0.95 for 95%).

    Returns:
        A tuple containing:
        - The sample mean.
        - The lower limit of the confidence interval.
        - The upper limit of the confidence interval.
    """
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
    sample_size = len(data)

    # Calculate the margin of error
    margin_of_error = stats.t.ppf((1 + confidence_level) / 2, sample_size - 1) * (sample_std / np.sqrt(sample_size))

    # Calculate the confidence interval limits
    lower_limit = sample_mean - margin_of_error
    upper_limit = sample_mean + margin_of_error

    return sample_mean, lower_limit, upper_limit

# Example usage:
sample_data = [23, 25, 28, 30, 32, 35, 37, 39]
sample_mean, lower_limit, upper_limit = confidence_interval(sample_data)

print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_limit:.2f}, {upper_limit:.2f})")
"""Explanation of Significance:
Confidence Interval:
The confidence interval provides a range of values within which the true population
mean is likely to fall, with a certain level of confidence.
It's not a probability statement about the population mean itself but rather
 a probability statement about the range.
Confidence Level:
The confidence level (e.g., 95%) specifies the probability that the interval
 will contain the true population mean if the sampling process were repeated many times.
Margin of Error:
The margin of error determines the width of the confidence interval.
It represents the maximum expected difference between the sample mean and
 the population mean.
```

```python
#ans10
import numpy as np
import scipy.stats as st

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level using sample data.

    Args:
        data (list or numpy.ndarray): Sample data.
        confidence (float, optional): Confidence level (e.g., 0.95 for 95% confidence).
            Defaults to 0.95.

    Returns:
        float: Margin of error.
    """
    data = np.array(data)
    n = len(data)
    mean = np.mean(data)
    std_err = st.sem(data)
    critical_value = st.t.ppf((1 + confidence) / 2, n - 1)
    margin_of_error = critical_value * std_err
    return margin_of_error

if __name__ == '__main__':
    sample_data = [155, 160, 165, 170, 175, 180, 185, 190, 195, 200]
    confidence_level = 0.95

    margin_error = calculate_margin_of_error(sample_data, confidence_level)
    print(f"Margin of error for {confidence_level*100}% confidence level: {margin_error:.2f}")

    confidence_interval = (np.mean(sample_data) - margin_error, np.mean(sample_data) + margin_error)
    print(f"Confidence interval: {confidence_interval}")
```

```python
#ans11
# Implementation of Bayesian Inference using Bayes' Theorem

import numpy as np

def bayes_theorem(prior_prob_a, prob_b_given_a, prob_b):
    """
    Calculates the posterior probability of event A given event B using Bayes' Theorem.

    Args:
        prior_prob_a (float): The prior probability of event A (P(A)).
        prob_b_given_a (float): The likelihood of event B given event A (P(B|A)).
        prob_b (float): The probability of event B (P(B)).

    Returns:
        float: The posterior probability of event A given event B (P(A|B)).
               Returns None if prob_b is zero to avoid division by zero.
    """
    if prob_b == 0:
        print("Error: Probability of evidence P(B) is zero. Cannot compute posterior.")
        return None

    # Bayes' Theorem: P(A|B) = [P(B|A) * P(A)] / P(B)
    posterior_prob_a_given_b = (prob_b_given_a * prior_prob_a) / prob_b

    return posterior_prob_a_given_b

# --- Explanation of the Process ---

# 1. Define the Problem:
#    Bayes' Theorem is used to update the probability of a hypothesis (Event A)
#    based on new evidence (Event B). We start with a 'prior' belief about the
#    probability of A, and update it to a 'posterior' probability after observing B.

#    Let's consider a medical example:
#    Event A: A person has a specific disease.
#    Event B: A person tests positive for the disease.

# 2. Establish Prior Probability (P(A)):
#    This is our initial belief about the probability of Event A occurring BEFORE
#    we consider the new evidence.
#    Example: Based on population statistics, the prevalence of the disease is 1%.
prior_prob_disease = 0.01 # P(Disease)

# 3. Determine Likelihood (P(B|A)):
#    This is the probability of observing the evidence (Event B) GIVEN that
#    the hypothesis (Event A) is true. This is often related to the accuracy
#    of a test or measurement.
#    Example: The diagnostic test has a sensitivity of 95% (correctly identifies
#    the disease when it's present).
prob_positive_given_disease = 0.95 # P(Positive Test | Disease)

# 4. Determine Probability of Evidence (P(B)):
#    This is the overall probability of observing the evidence (Event B),
#    regardless of whether Event A is true or not. This can be calculated
#    using the law of total probability:
#    P(B) = P(B|A) * P(A) + P(B|not A) * P(not A)
#    We need P(B|not A) (false positive rate) and P(not A) (1 - P(A)).
#    Example: The test has a specificity of 90% (correctly identifies when the
#    disease is NOT present). This means the false positive rate is 1 - specificity.
prob_positive_given_no_disease = 1 - 0.90 # P(Positive Test | No Disease) = 0.10
prior_prob_no_disease = 1 - prior_prob_disease # P(No Disease) = 0.99

prob_positive_test = (prob_positive_given_disease * prior_prob_disease) + \
                     (prob_positive_given_no_disease * prior_prob_no_disease) # P(Positive Test)

print(f"Prior probability of disease P(A): {prior_prob_disease}")
print(f"Likelihood P(B|A) (Test positive | Disease): {prob_positive_given_disease}")
print(f"False positive rate P(B|not A) (Test positive | No Disease): {prob_positive_given_no_disease}")
print(f"Prior probability of no disease P(not A): {prior_prob_no_disease}")
print(f"Probability of evidence P(B) (Overall positive test): {prob_positive_test:.4f}")
print("-" * 30)


```

```python
#ans12#ans12
# Chi-Square Test for Independence

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

def chi_square_test_independence(data, variable1, variable2, alpha=0.05):
    """
    Performs a Chi-square test for independence between two categorical variables.

    Args:
        data (pd.DataFrame): The DataFrame containing the categorical variables.
        variable1 (str): The name of the first categorical variable column.
        variable2 (str): The name of the second categorical variable column.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing:
            - pd.DataFrame: The contingency table.
            - float: The Chi-square statistic.
            - float: The p-value.
            - int: The degrees of freedom.
            - np.ndarray: The expected frequencies.
            - str: The interpretation of the results.
    """
    # Create a contingency table
    contingency_table = pd.crosstab(data[variable1], data[variable2])

    # Perform the Chi-square test
    chi2_statistic, p_value, degrees_of_freedom, expected_frequencies = chi2_contingency(contingency_table)

    # Interpret the results
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis (p < {alpha}). There is a statistically significant association between '{variable1}' and '{variable2}'."
    else:
        interpretation = f"Fail to reject the null hypothesis (p >= {alpha}). There is not enough evidence to suggest a statistically significant association between '{variable1}' and '{variable2}'."

    return contingency_table, chi2_statistic, p_value, degrees_of_freedom, expected_frequencies, interpretation

# Example Usage:
# Create a sample DataFrame with two categorical variables
data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female',
                   'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
        'Preference': ['A', 'B', 'A', 'A', 'B', 'B', 'A', 'B', 'A', 'A',
                       'B', 'B', 'A', 'B', 'A', 'A', 'B', 'B', 'A', 'B']}
df = pd.DataFrame(data)

# Perform the Chi-square test
contingency_table, chi2_stat, p_val, dof, expected_freq, interpretation = chi_square_test_independence(
    df, 'Gender', 'Preference'
)

print("Contingency Table:")
print(contingency_table)
print("-" * 30)
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(f"Degrees of Freedom: {dof}")
print("\nExpected Frequencies:")
print(expected_freq)
print("\nInterpretation:")
print(interpretation)

# Another Example:
data2 = {'Education': ['High School', 'College', 'High School', 'College', 'College', 'High School', 'College', 'High School'],
         'Job Satisfaction': ['High', 'High', 'Low', 'High', 'Low', 'Low', 'High', 'Low']}
df2 = pd.DataFrame(data2)

print("\n" + "=" * 30 + "\n")
contingency_table2, chi2_stat2, p_val2, dof2, expected_freq2, interpretation2 = chi_square_test_independence(
    df2, 'Education', 'Job Satisfaction'
)

print("Contingency Table (Example 2):")
print(contingency_table2)
print("-" * 30)
print(f"Chi-square Statistic: {chi2_stat2:.4f}")
print(f"P-value: {p_val2:.4f}")
print(f"Degrees of Freedom: {dof2}")
print("\nExpected Frequencies:")
print(expected_freq2)
print("\nInterpretation:")
print(interpretation2)
```

```python
#ans13
# Calculate Expected Frequencies for Chi-square Test

import numpy as np
import pandas as pd

def calculate_expected_frequencies(observed_data):
    """
    Calculates the expected frequencies for a Chi-square test of independence
    based on a contingency table of observed data.

    Args:
        observed_data (pd.DataFrame or np.ndarray): A contingency table
            (cross-tabulation) of observed frequencies. Rows and columns
            represent the categories of the two variables.

    Returns:
        pd.DataFrame or np.ndarray: The calculated expected frequencies.
    """
    if isinstance(observed_data, np.ndarray):
        observed_df = pd.DataFrame(observed_data)
    elif isinstance(observed_data, pd.DataFrame):
        observed_df = observed_data
    else:
        raise TypeError("observed_data must be a pandas DataFrame or a numpy array.")

    row_totals = observed_df.sum(axis=1)
    col_totals = observed_df.sum(axis=0)
    grand_total = observed_df.sum().sum()

    # Handle the case where grand_total is zero to avoid division by zero
    if grand_total == 0:
        print("Warning: Grand total is zero. Expected frequencies will be zero.")
        return np.zeros_like(observed_df)

    expected_frequencies = np.outer(row_totals, col_totals) / grand_total

    # Return as DataFrame if input was DataFrame, otherwise return as numpy array
    if isinstance(observed_data, pd.DataFrame):
        return pd.DataFrame(expected_frequencies, index=observed_df.index, columns=observed_df.columns)
    else:
        return expected_frequencies


# Example Usage:

# Example 1: Using a pandas DataFrame as input (from a previous Chi-square test)
# Let's use the contingency table from the previous Chi-square test example:
data = {'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female',
                   'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
        'Preference': ['A', 'B', 'A', 'A', 'B', 'B', 'A', 'B', 'A', 'A',
                       'B', 'B', 'A', 'B', 'A', 'A', 'B', 'B', 'A', 'B']}
df = pd.DataFrame(data)
observed_contingency_table_df = pd.crosstab(df['Gender'], df['Preference'])

print("Observed Contingency Table (DataFrame):")
print(observed_contingency_table_df)
print("-" * 30)

expected_freq_df = calculate_expected_frequencies(observed_contingency_table_df)

print("Calculated Expected Frequencies (from DataFrame):")
print(expected_freq_df)
print("-" * 30)


# Example 2: Using a NumPy array as input
# A 2x2 contingency table represented as a NumPy array
observed_array = np.array([[10, 20], [30, 40]])

print("Observed Contingency Table (NumPy Array):")
print(observed_array)
print("-" * 30)

expected_freq_array = calculate_expected_frequencies(observed_array)

print("Calculated Expected Frequencies (from NumPy Array):")
print(expected_freq_array)

# Note: When performing a full Chi-square test using scipy.stats.chi2_contingency,
# the expected frequencies are already calculated and returned as part of the output.
# This function is useful if you specifically need to calculate expected frequencies
# from a given observed contingency table without running the full test. [1]
```

```python
#ans14
# Chi-Square Goodness-of-Fit Test

import numpy as np
from scipy.stats import chisquare

def chi_square_goodness_of_fit(observed_counts, expected_counts, alpha=0.05):
    """
    Performs a Chi-square goodness-of-fit test.

    This test determines if a sample of observed data comes from a population
    with a specific hypothesized distribution.

    Args:
        observed_counts (array-like): The observed frequencies (counts)
                                     for each category.
        expected_counts (array-like): The expected frequencies (counts)
                                     for each category under the null hypothesis.
                                     Must have the same length as observed_counts.
        alpha (float, optional): The significance level. Defaults to 0.05.

    Returns:
        tuple: A tuple containing:
            - float: The Chi-square statistic.
            - float: The p-value.
            - str: The interpretation of the results.
    """
    # Ensure observed and expected counts have the same length
    if len(observed_counts) != len(expected_counts):
        raise ValueError("Observed counts and expected counts must have the same length.")

    # Ensure expected counts are not zero in any category (can cause issues with test)
    if np.any(np.array(expected_counts) <= 0):
         print("Warning: Some expected counts are zero or negative. Ensure expected counts are positive.")
         # chisquare handles expected counts of 0 gracefully by excluding the category,
         # but small expected counts (e.g., < 5) can make the test less reliable.

    # Perform the Chi-square goodness-of-fit test
    # The chisquare function calculates the statistic and p-value.
    # Degrees of freedom is typically (number of categories - 1)
    # If parameters of the distribution were estimated from the data,
    # df is (number of categories - 1 - number of estimated parameters).
    # chisquare automatically calculates df as (number of categories - 1)
    # unless f_exp (expected_counts) implies constraints.
    chi2_statistic, p_value = chisquare(f_obs=observed_counts, f_exp=expected_counts)

    # Interpret the results
    # The null hypothesis (H0) is that the observed data fits the expected distribution.
    # The alternative hypothesis (H1) is that the observed data does not fit the expected distribution.
    if p_value < alpha:
        interpretation = f"Reject the null hypothesis (p < {alpha}). The observed data does not fit the expected distribution."
    else:
        interpretation = f"Fail to reject the null hypothesis (p >= {alpha}). The observed data fits the expected distribution."

    return chi2_statistic, p_value, interpretation

# Example Usage:

# Scenario: We roll a standard six-sided die 120 times and record the results.
# We want to test if the die is fair (i.e., if the outcomes follow a uniform distribution).

# Expected distribution for a fair die rolled 120 times:
# Each side (1, 2, 3, 4, 5, 6) is expected to appear 120 / 6 = 20 times.
expected_counts = [20, 20, 20, 20, 20, 20]

# Observed results from rolling the die 120 times:
# Let's assume these are the counts we observed for each side.
observed_counts = [15, 25, 18, 22, 19, 21] # Sum = 120

# Significance level
alpha = 0.05

# Perform the goodness-of-fit test
chi2_stat, p_val, interpretation = chi_square_goodness_of_fit(observed_counts, expected_counts, alpha)

print("Observed Counts:", observed_counts)
print("Expected Counts:", expected_counts)
print("-" * 30)
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print("\nInterpretation:")
print(interpretation)


# Another Example: Testing if observed data fits a specific ratio (e.g., Mendelian genetics 3:1 ratio)
# Scenario: Observe offspring from a genetic
```
