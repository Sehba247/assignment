```python
#ans3
import matplotlib.pyplot as plt
import pandas as pd

# Load the data
price_df=pd.read_csv('/amazon (2).csv')

# Clean the price columns: remove '₹' and ',' and convert to float
price_df['discounted_price_clean'] = price_df['discounted_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)
price_df['actual_price_clean'] = price_df['actual_price'].astype(str).str.replace('₹', '').str.replace(',', '').astype(float)


# Plot distribution using boxplots for cleaned prices
plt.figure(figsize=(10, 6))
plt.boxplot(
    [price_df['discounted_price_clean'], price_df['actual_price_clean']],
    labels=['Discounted Price', 'Actual Price'],
    patch_artist=True,
    boxprops=dict(facecolor='lightblue')
)

plt.title('Distribution of Discounted Prices vs. Actual Prices')
plt.ylabel('Price (₹)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#ans3The boxplot above compares the distribution of discounted and actual prices:

#Key Observations:
#Actual prices tend to be significantly higher, as expected.

#There are many high-value outliers, especially in actual prices, reflecting premium or luxury products.

#The median discounted price is noticeably lower than the median actual price, indicating effective discounting.
```
